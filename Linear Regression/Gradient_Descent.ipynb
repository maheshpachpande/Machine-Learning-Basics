```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=65, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# lets apply ols
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# lets apply Gradient descent assuming slope is constant m = 75.07
# and lets assume starting intercept for b = 0
y_pred = ((75.07*x)+0).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='ols')
plt.plot(x,y_pred,color='black',label='b=0')
plt.legend()
plt.show()
```

```python
m = 75.07
b = 0
```

```python
loss_slope = -2 * np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
lr=0.1
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((75.07*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='ols')
plt.plot(x,y_pred1,color='black',label='b={}'.format(b))
plt.plot(x,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m * x.ravel() - b)
loss_slope
```

```python
step_size = lr * loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((75.07*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='ols')
plt.plot(x,y_pred1,color='black',label='b={}'.format(b))
plt.plot(x,y_pred2,color='black',label='b={}'.format(b))
plt.plot(x,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m * x.ravel() - b)
step_size = lr * loss_slope
b = b - step_size
y_pred3 = ((75.07*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='ols')
plt.plot(x,y_pred1,color='#00a65a',label='b={}'.format(b))
plt.plot(x,y_pred2,color='#00a65a',label='b={}'.format(b))
plt.plot(x,y_pred3,color='black',label='b={}'.format(b))
plt.plot(x,y_pred,color='black',label='b=0')
plt.legend()
plt.show()
```

```python
b = -110
m = 75.07
lr = 0.1
epochs = 10
for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (loss_slope*lr)
  y_pred = m*x + b
  plt.plot(x,y_pred)
plt.scatter(x,y)
```

```python
# small learnimg rate
b = -110
m = 75.07
lr = 0.01
epochs = 100
for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*x.ravel() - b)
  b = b - (loss_slope*lr)
  y_pred = m*x + b
  plt.plot(x,y_pred)
plt.scatter(x,y)
```

```python
class GDRegression:
  def __init__(self,learning_rate,epochs):
    self.m = 28.32
    self.b = -120
    self.epochs = epochs
    self.lr = learning_rate

  def fit(self,x,y):
    for i in range(self.epochs):
      loss_slope = -2 * np.sum(y - self.m*x.ravel() - self.b)
      self.b = self.b - (loss_slope*self.lr)
      print(loss_slope,self.b)
    print(self.b)
```
