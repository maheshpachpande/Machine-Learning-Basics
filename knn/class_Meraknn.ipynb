```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(df.iloc[:,0:3].values)
y = df.iloc[:,-1].values

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=13)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
from sklearn.metrics import accuracy_score
print('accuracy score : ', accuracy_score(y_test, y_pred))
```

```python
from collections import Counter
```

```python
class MeraKnn:
  def __init__(self,n_neighbors=5):
    # initiate variables like k, x_train, x_test
    self.n_neighbors = n_neighbors
    self.x_train = None
    self.y_train = None

  def fit(self,x_train,y_train):
    # knn is lazy learner so that only assigning here
    self.x_train = x_train
    self.y_train = y_train

  def predict(self,x_test):
    # make prediction
    y_pred = []
    for i in x_test:
      # calculate distance between each x_test and training point
      distances=[]
      for j in x_train:
        distances.append(self.calculate_distance(i,j))
      # getting neighbors of test point
      n_neighbors = sorted(list(enumerate(distances)), key = lambda x : x[1])[0:self.n_neighbors]
      # get label of nearest neighbors
      label = self.majority_count(n_neighbors)
      y_pred.append(label)
    return np.array(y_pred)


  def calculate_distance(self,point_A, point_B):
    '''calculate distance between x_test and x_train'''
    return np.linalg.norm(point_A - point_B)


  def majority_count(self,n_neighbors):
    '''count majority of calculated nearest neighbhors'''
    # vote bank
    votes = []

    for i in n_neighbors:
      # catch y_train value of test point
      votes.append(self.y_train[i[0]])

    # get counter objects
    votes = Counter(votes)
    # get most common
    return votes.most_common()[0][0]
```

```python
knn = MeraKnn(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
print('accuracy score : ', accuracy_score(y_test, y_pred))
```
